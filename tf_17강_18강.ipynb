```
import os
import json
import numpy as np

import tensorflow as tf
import tensorflow_datasets as tfds

from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Flatten, Dense, Activation
from tensorflow.keras.layers import InputLayer

os.environ['TF_FORCE_GPU_ALLOW_GROWTH'] = 'true'
```

```
model = Sequential()
model.add(Dense(units=10))
model.add(Activation('relu'))
model.add(Dense(units=2))
model.add(Activation('softmax'))

print(model.summary()) # input이 정의되지 않으면 error, build를 해주어야함
```

```
model = Sequential()
# model.add(InputLayer(input_shape=(28, 28, 1)))
model.add(Flatten())
model.add(Dense(units=10))
model.add(Activation('relu'))
model.add(Dense(units=2))
model.add(Activation('softmax'))

# model.build()
model.build(input_shape = (None, 28, 28, 1)) # batchsize, 28, 28, channel -> None: Batchsize는 그때 그때 달라짐
model.summary()

# weight와 bias가 몇개가 들어있는지 확인해보자
# tf.keras.backend.clear_session()
```

```
class TestModel(Model):
  def __init__(self):
    super(TestModel, self).__init__()

    self.flatten = Flatten(name = 'fffffffffflatten')
    self.d1 = Dense(units=10, name = 'dddddddddddddense')
    self.d1_act = Activation('relu')
    self.d2 = Dense(units=2)
    self.d2_act = Activation('softmax')

  def call(self, x):
    x = self.flatten(x)
    x = self.d1(x)
    x = self.d1_act(x)
    x = self.d2(x)
    x = self.d2_act(x)
    return x

model = TestModel()
if not model.built:
  model.build(input_shape = (None, 28, 28, 1))

# 굳이 build를 해주지 않더라도 input을 넣게 되면 그에 맞게 build가 된다.

model.summary()
```
